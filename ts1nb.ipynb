```python
amp=1
freq=1001
offset=np.pi
fs=1000
N=1000
ts=1/fs
```

```python
tt = np.linspace(0, (N-1)*ts, N)

sin_sig = amp * np.sin(2*np.pi*freq*tt+offset)
```

```python
plt.plot(tt, sin_sig)
plt.xlabel('Grilla temporal (tt)')
plt.ylabel('Amplitud sin_sig(tt)')
plt.axis('tight')
plt.show()
```
